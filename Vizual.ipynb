```python
from pylab import rcParams
rcParams['figure.figsize'] = 10,7

import pandas as pd
import matplotlib.pyplot as plt
```

```python
rating = pd.read_csv('IMDB-Movie-Data.csv')
```

```python
rating.info()
```

```python
rating_top=rating.groupby('Director').mean().sort_values('Rating', ascending=False).head(15)
rating_top=rating_top.reset_index(level=0, drop=False)
rating_top.head(15)
```

```python
plt.barh(rating_top['Director'], rating_top['Rating'])
plt.xlabel('Рейтинг')
plt.ylabel('Режиссеры')
plt.title('Топ 15 режиссеров')
plt.text(10, 1, 'Самые востребованные режиссеры')
plt.grid(15)
```

```python
name_dynamics = rating.groupby(['Director']).sum().sort_values(by='Votes', ascending=False).head(5)
name_dynamics.plot(kind='pie', y='Votes')
```

```python
df = pd.read_csv('IMDB-Movie-Data.csv')
print(df.shape)
```

```python
df.info()
df = df.dropna()
print(df.shape)
```

```python
df.head()
```

```python
df['Rating'] = df.Rating.astype('float64')
df['Year'] = df.Year.astype('int64')
df['Votes'] = df.Votes.astype('int64')
df['Metascore'] = df.Metascore.astype('int64')
```

```python
[x for x in df.columns if 'Revenue' in x]
```

```python
df1 = df[[x for x in df.columns if 'Revenue' in x] + ['Year']]\
    .groupby('Year').sum()
df1.head()
```

```python
df1.plot()
```

```python
df1.plot(kind='bar', rot=45)
```

```python
ax = df.Rating.hist()
ax.set_title('Успешность фильма')
ax.set_xlabel('Рейтинг')
ax.set_ylabel('')
```
